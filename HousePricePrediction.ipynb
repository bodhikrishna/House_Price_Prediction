```python
def isfloat(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
def convert_to_float(x):
  inputs=x.split('-')
  if len(inputs)==2:
    return (float(inputs[0])+float(inputs[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python

# for other sq feet values coonversion use these functions to convert all possible ratios to sq feet.
# These are not necessary , depends on user to adopt
def convert_from_SqMeter(x):
  import re
  num = re.findall(r'[-+]?\d*\.\d+|\d+', x)
  if num:
    sqnum=float(num[0])


  sqfeet=sqnum*10.7639
  return sqfeet

def convert_from_SqYards(x):
  import re
  num = re.findall(r'[-+]?\d*\.\d+|\d+', x)
  if num:
    sqnum=float(num[0])


  sqyards=sqnum*9
  return sqyards

def convert_from_acres(x):
  import re
  num = re.findall(r'[-+]?\d*\.\d+|\d+', x)
  if num:
    sqnum=float(num[0])


  sqAcres=sqnum*43560
  return sqAcres
```

```python
def remove_pricePerSqft_outliers(df):
  df_out=pd.DataFrame()
  for key , subdf in df.groupby('location'):
    mean=np.mean(subdf.price_per_sqfeet)
    SD=np.std(subdf.price_per_sqfeet)
    reduced_df=subdf[(subdf.price_per_sqfeet>(mean-SD)) & (subdf.price_per_sqfeet<=(mean+SD))]
    df_out= pd.concat([df_out,reduced_df],ignore_index=True)
  return df_out

df7=remove_pricePerSqft_outliers(df6)
df7.shape

```

```python
def Scatter_plot_chart(df,location):
  bhk2=df[(df.location==location) & (df.bhk==2)]
  bhk3=df[(df.location==location) & (df.bhk==3)]
  matplotlib.rcParams['figure.figsize']=(15,10)
  plt.scatter(bhk2.total_sqft,bhk2.price,color="blue", label="2 BHK", s=20)
  plt.scatter(bhk3.total_sqft,bhk3.price,color="green", label="3 BHK", s=20)
  plt.xlabel("Total Sq Feet Area")
  plt.ylabel("Price")
  plt.legend()


Scatter_plot_chart(df7,"Rajaji Nagar")
```

```python
def remove_bhk_outliers(df):
  exclude_indexes=np.array([])
  for location,location_df in df.groupby('location'):
    bhk_stats={}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_stats[bhk]={
          'mean': np.mean(bhk_df.price_per_sqfeet),
          'std': np.std(bhk_df.price_per_sqfeet),
          'count': bhk_df.shape[0]
      }
    for bhk , bhk_df in location_df.groupby('bhk'):
      stats=bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indexes=np.append(exclude_indexes, bhk_df[bhk_df.price_per_sqfeet<(stats['mean'])].index.values)
  return df.drop(exclude_indexes,axis='index')
```

```python
from sklearn.model_selection import ShuffleSplit, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# def best_model_using_gridsearchCV(X,y):
#   algos={
#       'linear_regression':{
#           'model':LinearRegression(),
#           'params':{
#               'fit_intercept':[True, False]
#           }
#       },
#       'lasso':{
#           'model':Lasso(),
#           'params':{
#               'alpha':[1,2],
#               'selection':['random','cyclic']
#           }
#       },
#       'decision_tree':{
#           'model':DecisionTreeRegressor(),
#           'params': {
#               'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
#               'splitter':['best','random']
#           }
#       }

#       }


#   scores=[]
#   cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
#   for algo_name, config in algos.items():
#     gs= GridSearchCV(config['model'],config['params'],cv=cv, return_train_score=False)
#     gs.fit(X,Y)
#     scores.append({
#         'model':algo_name,
#         'best_score':gs.best_score_,
#         'best_params':gs.best_params_
#     })
#   return pd.DataFrame(scores,columns=['model','best_score','best_params'])
def best_model_using_gridsearchCV(X, y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
         'lasso':{
          'model':Lasso(),
          'params':{
              'alpha':[1,2],
              'selection':['random','cyclic']
          }
      },
      'decision_tree':{
          'model':DecisionTreeRegressor(),
          'params': {
              'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
              'splitter':['best','random']
          }
      }

    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X, y)

        if algo_name == 'linear_regression':
            y_pred = gs.best_estimator_.predict(X)
            mse = mean_squared_error(y, y_pred)
            mae = mean_absolute_error(y, y_pred)
            r2 = r2_score(y, y_pred)

            scores.append({
                'model': algo_name,
                'best_score': gs.best_score_,
                'best_params': gs.best_params_,
                'mean_squared_error': mse,
                'mean_absolute_error': mae,
                'r2_score': r2
            })
        else:
            scores.append({
                'model': algo_name,
                'best_score': gs.best_score_,
                'best_params': gs.best_params_
            })

    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params', 'mean_squared_error', 'mean_absolute_error', 'r2_score'])


best_model_using_gridsearchCV(X,Y)
```

```python
def predictPrice(location,sqft,bath,bhk):
  loc_index=np.where(X.columns==location)[0][0]
  x=np.zeros(len(X.columns))
  x[0]=sqft
  x[1]=bath
  x[2]=bhk
  if loc_index >=0:
    x[loc_index]=1

  return lr_model.predict([x])[0]
```
